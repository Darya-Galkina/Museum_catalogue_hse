```python
import os
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler  #  
from torchvision import transforms, models
import pandas as pd
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    average_precision_score,
    roc_curve,
    auc,
    precision_recall_curve,
    confusion_matrix
)

# 
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
BATCH_SIZE = 16
IMG_SIZE = 224
PATIENCE = 5
CLASS_THRESHOLD = 0.4
POS_WEIGHT = 3.0

class ArtworkDataset(Dataset):
    def __init__(self, df):
        self.df = df.reset_index(drop=True)

        self.transform_sym = transforms.Compose([
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomVerticalFlip(p=0.3),
            transforms.Resize(IMG_SIZE),
            transforms.CenterCrop(IMG_SIZE),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ])

        self.transform_asym = transforms.Compose([
            transforms.RandomAffine(degrees=0, shear=20),
            transforms.RandomPerspective(distortion_scale=0.6, p=0.8),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomVerticalFlip(p=0.5),
            transforms.ColorJitter(0.2, 0.2, 0.2),
            transforms.GaussianBlur(3),
            transforms.Resize(IMG_SIZE),
            transforms.CenterCrop(IMG_SIZE),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ])

        self.valid_indices = []
        for idx in range(len(self.df)):
            path = self.df.iloc[idx]['path']
            if os.path.exists(path):
                self.valid_indices.append(idx)
            else:
                print(f"  : {path}")

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        real_idx = self.valid_indices[idx]
        row = self.df.iloc[real_idx]

        try:
            with Image.open(row['path']) as img:
                label = 0 if row['label'] == "" else 1
                transform = self.transform_sym if label == 0 else self.transform_asym
                image = transform(img.convert('RGB'))
                return image, label
        except Exception as e:
            print(f"  {row['path']}: {str(e)}")
            return torch.zeros((3, IMG_SIZE, IMG_SIZE)), -1

class EnhancedSymmetryClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = models.efficientnet_b7(pretrained=True)
        in_features = self.model.classifier[1].in_features

        self.model.classifier = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(in_features, 1)
        )

    def forward(self, x):
        return self.model(x)

def calculate_metrics(model, loader):
    model.eval()
    all_probs = []
    all_preds = []
    all_labels = []
    confidences = []

    with torch.no_grad():
        for images, labels in loader:
            if -1 in labels:
                continue

            images = images.to(device)
            outputs = model(images).squeeze()

            probs = torch.sigmoid(outputs).cpu().numpy()
            preds = (probs > CLASS_THRESHOLD).astype(int)
            confidence = np.abs(probs - 0.5) * 2

            all_probs.extend(probs)
            all_preds.extend(preds)
            all_labels.extend(labels.numpy())
            confidences.extend(confidence)

    metrics = {
        'accuracy': accuracy_score(all_labels, all_preds),
        'precision_class_0': precision_score(all_labels, all_preds, pos_label=0),
        'precision_class_1': precision_score(all_labels, all_preds, pos_label=1),
        'recall_class_0': recall_score(all_labels, all_preds, pos_label=0),
        'recall_class_1': recall_score(all_labels, all_preds, pos_label=1),
        'f1_class_0': f1_score(all_labels, all_preds, pos_label=0),
        'f1_class_1': f1_score(all_labels, all_preds, pos_label=1),
        'roc_auc': roc_auc_score(all_labels, all_probs),
        'pr_auc': average_precision_score(all_labels, all_probs),
        'confusion_matrix': confusion_matrix(all_labels, all_preds),
        'avg_confidence': np.mean(confidences)
    }

    return metrics, all_probs, all_labels, confidences

def visualize_metrics(metrics, probs, labels, confidences):
    print("\n" + "="*50)
    print(f"   : {CLASS_THRESHOLD}")
    print(f"   1: {POS_WEIGHT}")
    print("\n  :")
    print(f" (Accuracy): {metrics['accuracy']:.4f}")
    print(f"ROC-AUC: {metrics['roc_auc']:.4f}")
    print(f"PR-AUC: {metrics['pr_auc']:.4f}")
    print(f" : {metrics['avg_confidence']:.4f}")

    print("\n  :")
    print(f" '' (0):")
    print(f"  Precision: {metrics['precision_class_0']:.4f}")
    print(f"  Recall: {metrics['recall_class_0']:.4f}")
    print(f"  F1-score: {metrics['f1_class_0']:.4f}")

    print(f"\n '' (1):")
    print(f"  Precision: {metrics['precision_class_1']:.4f}")
    print(f"  Recall: {metrics['recall_class_1']:.4f}")
    print(f"  F1-score: {metrics['f1_class_1']:.4f}")

    plt.figure(figsize=(10, 6))
    plt.hist(confidences, bins=20, range=(0, 1))
    plt.title('  ')
    plt.xlabel('')
    plt.ylabel(' ')
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.heatmap(metrics['confusion_matrix'],
                annot=True, fmt='d',
                cmap='Blues',
                xticklabels=['', ''],
                yticklabels=['', ''])
    plt.title(' ')
    plt.xlabel(' ')
    plt.ylabel(' ')
    plt.show()

    fpr, tpr, _ = roc_curve(labels, probs)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(10, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2,
             label=f'ROC  (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-')
    plt.legend(loc="lower right")
    plt.show()

    precision, recall, _ = precision_recall_curve(labels, probs)
    average_precision = average_precision_score(labels, probs)
    plt.figure(figsize=(10, 6))
    plt.step(recall, precision, where='post',
             label=f'PR  (AP = {average_precision:.2f})')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title('Precision-Recall ')
    plt.legend(loc="upper right")
    plt.show()

def create_loaders(df, batch_size):
    df = df[df['path'].apply(os.path.exists)]

    #    
    class_counts = df['label'].value_counts()
    class_weights = 1. / class_counts
    sample_weights = [class_weights[label] for label in df['label']]

    #  WeightedRandomSampler  
    sampler = WeightedRandomSampler(
        weights=sample_weights,
        num_samples=len(sample_weights),
        replacement=True
    )

    dataset = ArtworkDataset(df)
    loader = DataLoader(
        dataset,
        batch_size=batch_size,
        sampler=sampler,
        num_workers=0,
        pin_memory=False
    )
    return loader

def train_model():
    df = pd.read_csv('G:/project/symmetry.csv', header=None, names=['path', 'label'])
    df = df[df['label'].isin(['', ''])]

    train_df = df.sample(frac=0.8, random_state=42)
    val_df = df.drop(train_df.index)

    train_loader = create_loaders(train_df, BATCH_SIZE)
    val_loader = create_loaders(val_df, BATCH_SIZE)

    model = EnhancedSymmetryClassifier().to(device)
    criterion = nn.BCEWithLogitsLoss(pos_weight=torch.tensor([POS_WEIGHT]).to(device))
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'max', patience=2)

    best_roc_auc = 0
    no_improve = 0
    best_model_weights = None

    train_losses = []
    val_metrics_history = []

    for epoch in range(30):
        model.train()
        epoch_loss = 0
        for images, labels in train_loader:
            valid_mask = labels != -1
            if not valid_mask.any():
                continue

            images = images[valid_mask].to(device)
            labels = labels[valid_mask].float().to(device)

            optimizer.zero_grad()
            outputs = model(images).squeeze()
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()

        avg_train_loss = epoch_loss / len(train_loader)
        train_losses.append(avg_train_loss)

        metrics, probs, labels, confidences = calculate_metrics(model, val_loader)
        val_metrics_history.append(metrics)
        scheduler.step(metrics['roc_auc'])

        print(f"\nEpoch {epoch+1} :")
        print(f"Train Loss: {avg_train_loss:.4f}")
        visualize_metrics(metrics, probs, labels, confidences)

        if metrics['roc_auc'] > best_roc_auc:
            best_roc_auc = metrics['roc_auc']
            no_improve = 0
            best_model_weights = model.state_dict()
            torch.save(best_model_weights, 'best_model.pth')
            print(f"    (ROC-AUC: {best_roc_auc:.4f})")
        else:
            no_improve += 1
            print(f"  {no_improve}/{PATIENCE}")
            if no_improve >= PATIENCE:
                print(f"    {epoch+1}")
                break

    model.load_state_dict(best_model_weights)
    final_metrics, final_probs, final_labels, final_conf = calculate_metrics(model, val_loader)
    print("\n :")
    visualize_metrics(final_metrics, final_probs, final_labels, final_conf)

    plt.figure(figsize=(10, 6))
    plt.plot(train_losses, label='Training Loss')
    plt.plot([m['roc_auc'] for m in val_metrics_history], label='Validation ROC-AUC')
    plt.title(' ')
    plt.xlabel('')
    plt.ylabel('')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    torch.multiprocessing.freeze_support()
    train_model()
```

```python
import os
import torch
import torch.nn as nn
from torchvision import transforms, models
from PIL import Image
import pandas as pd
import numpy as np
from torch.utils.data import Dataset, DataLoader

# 
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
IMG_SIZE = 224
PATIENCE = 7
CLASS_THRESHOLD = 0.3
POS_WEIGHT = 5.0

class SymmetryClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = models.efficientnet_b7(pretrained=False)
        in_features = self.model.classifier[1].in_features
        self.model.classifier = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(in_features, 1)
        )

    def forward(self, x):
        return self.model(x)

class InferenceDataset(Dataset):
    def __init__(self, image_paths):
        self.image_paths = image_paths
        self.transform = transforms.Compose([
            transforms.Resize(IMG_SIZE),
            transforms.CenterCrop(IMG_SIZE),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ])

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        path = self.image_paths[idx]
        try:
            with Image.open(path) as img:
                image = self.transform(img.convert('RGB'))
                return image, path
        except Exception as e:
            print(f"  {path}: {str(e)}")
            return torch.zeros((3, IMG_SIZE, IMG_SIZE)), path

def load_model(model_path):
    model = SymmetryClassifier().to(device)
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.eval()
    return model

def predict(model, test_csv_path, output_csv_path):
    #  
    df = pd.read_csv(test_csv_path)
    image_paths = df['image_path'].tolist()

    #   
    valid_paths = [p for p in image_paths if os.path.exists(p)]
    if len(valid_paths) != len(image_paths):
        print(f": {len(image_paths) - len(valid_paths)}   ")

    dataset = InferenceDataset(valid_paths)
    loader = DataLoader(dataset, batch_size=16, shuffle=False)

    # 
    results = []
    with torch.no_grad():
        for images, paths in loader:
            images = images.to(device)
            outputs = model(images).squeeze()
            probs = torch.sigmoid(outputs).cpu().numpy()

            for path, prob in zip(paths, probs):
                pred_class = 1 if prob > CLASS_THRESHOLD else 0
                confidence = abs(prob - 0.5) * 2
                results.append({
                    'image_path': path,
                    'prediction': '' if pred_class == 1 else '',
                    'probability': float(prob),
                    'confidence': float(confidence)
                })

    #  
    result_df = pd.DataFrame(results)
    result_df.to_csv(output_csv_path, index=False)
    print(f"   {output_csv_path}")
    return result_df

if __name__ == "__main__":
    #    (  )
    MODEL_PATH = 'best_model.pth'  #    
    TEST_CSV_PATH = 'sym_test.csv'  #     
    OUTPUT_CSV_PATH = 'predictions.csv'  #    

    #    
    model = load_model(MODEL_PATH)
    predictions = predict(model, TEST_CSV_PATH, OUTPUT_CSV_PATH)

    #   5 
    print("\n :")
    print(predictions.head())
```

```python
import pandas as pd
import os

def evaluate_predictions(predictions_csv):
    #  
    df = pd.read_csv(predictions_csv)

    #      
    df['true_label'] = df['image_path'].apply(
        lambda x: '' if 'asymmetry' in x.lower() else
                 ('' if 'symmetry' in x.lower() else None)
    )

    #  ,     
    df = df.dropna(subset=['true_label'])

    if len(df) == 0:
        print("    'asymmetry'  'symmetry'  ")
        return

    #  
    accuracy = (df['prediction'] == df['true_label']).mean()

    #   
    class_report = {}
    for class_name in ['', '']:
        class_data = df[df['true_label'] == class_name]
        if len(class_data) > 0:
            precision = (class_data['prediction'] == class_name).mean()
            recall = (class_data['prediction'] == class_name).mean()  #   
            class_report[class_name] = {
                'precision': precision,
                'recall': recall,
                'support': len(class_data)
            }

    #  
    print(f" : {accuracy:.4f}")
    print("\n  :")
    for class_name, metrics in class_report.items():
        print(f"\n '{class_name}':")
        print(f"  Precision: {metrics['precision']:.4f}")
        print(f"  Recall: {metrics['recall']:.4f}")
        print(f"   : {metrics['support']}")

    #  
    confusion = pd.crosstab(
        df['true_label'],
        df['prediction'],
        rownames=[''],
        colnames=['']
    )
    print("\n :")
    print(confusion)

    return df

if __name__ == "__main__":
    #     
    PREDICTIONS_CSV = 'predictions.csv'

    #  
    evaluated_data = evaluate_predictions(PREDICTIONS_CSV)

    #        
    evaluated_data.to_csv('evaluated_predictions.csv', index=False)
```
